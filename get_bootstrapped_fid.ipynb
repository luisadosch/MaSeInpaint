```python
def get_mean(data, string):
    # Mittelwert berechnen
    mean_value = np.mean(data)
    
    # Ausgabe des Mittelwerts mit String
    print(f"Bootstrapped Mittelwert {string}: {mean_value}")
    
    return mean_value
```

```python
def get_min_max(data):
    min = np.min(data)
    max = np.max(data)
    print("Min:", min)
    print("Max:", max)
    
    return min, max
    
```

```python
def plot_hist(data, title_string):
    plt.hist(data, bins=30)
    plt.title(f"Histogramm der Bootstrapped {title_string} FID")
    plt.xlabel("FID-Werte")
    plt.ylabel("Häufigkeit")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()
```

```python
plot_hist(fid_lama, "LaMa")
mean_value = get_mean(fid_lama, "LaMa auf ganzen Datensatz")
min, max = get_min_max(fid_lama)

df.loc[len(df)] = ['Lama', "Alle", mean_value, min, max]
```
