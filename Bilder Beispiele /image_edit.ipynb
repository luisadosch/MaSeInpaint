```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_original = mpimg.imread('original_removal.jpg')  # Replace with your first image file path
img_after = mpimg.imread('after_removal.jpg')  # Replace with your second image file path
```

```python
# Create a figure and axis to place the images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# Display images
ax[0].imshow(img_original)
ax[0].axis('off')  # Hide axis for the first image
ax[0].set_title('Original Image')  # Add caption for the first image

ax[1].imshow(img_after)
ax[1].axis('off')  # Hide axis for the second image
ax[1].set_title('After Object Removal')  # Add caption for the second image

# Adjust layout and save the plot
plt.tight_layout()  # Adjust layout to fit the images
plt.savefig('object_removal_comparison.jpg')
plt.show()
```

```python
before_denoising = mpimg.imread('before_denoising.jpg')  # Replace with your first image file path
after_denoising = mpimg.imread('before_denoising_processed.jpg')  # Replace with your second image file path
```

```python
# Create a figure and axis to place the images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# Display images
ax[0].imshow(before_denoising)
ax[0].axis('off')  # Hide axis for the first image
ax[0].set_title('Original Image')  # Add caption for the first image

ax[1].imshow(after_denoising)
ax[1].axis('off')  # Hide axis for the second image
ax[1].set_title('After Denoising')  # Add caption for the second image

# Show the plot
plt.tight_layout()  # Adjust layout to fit the images
#save the plot
plt.savefig('denoising_comparison.jpg')
plt.show()
```
